```python
import pandas as pd
import numpy as np

# Load cleaned dataset
df_fe = pd.read_csv("credit_scoring_dataset_cleaned.csv")

# -----------------------------
# 0️⃣ Fill missing values
# -----------------------------
df_fe['Mto_ingreso_mensual'] = df_fe['Mto_ingreso_mensual'].fillna(df_fe['Mto_ingreso_mensual'].median())
df_fe['Prct_deuda_vs_ingresos'] = df_fe['Prct_deuda_vs_ingresos'].fillna(df_fe['Prct_deuda_vs_ingresos'].median())
df_fe['Nro_dependiente'] = df_fe['Nro_dependiente'].fillna(0)
df_fe['Nro_prod_financieros_deuda'] = df_fe['Nro_prod_financieros_deuda'].fillna(0)
df_fe['Edad'] = df_fe['Edad'].fillna(df_fe['Edad'].median())

# -----------------------------
# 1️⃣ Income per person
# -----------------------------
# Represents the monthly income available per household member
df_fe['Income_per_person'] = df_fe['Mto_ingreso_mensual'] / (df_fe['Nro_dependiente'] + 1)
df_fe['Income_per_person'] = df_fe['Income_per_person'].replace([np.inf, -np.inf], np.nan).fillna(df_fe['Income_per_person'].median())

# -----------------------------
# 2️⃣ Total delays
# -----------------------------
# Sum of all delay/retraso columns → overall late payments
delay_cols = [col for col in df_fe.columns if 'retraso' in col.lower() or 'delay' in col.lower()]

if delay_cols:  # If delay columns exist
    df_fe['Total_delays'] = df_fe[delay_cols].sum(axis=1)
    df_fe['Total_delays_norm'] = df_fe['Total_delays'] / (df_fe['Total_delays'].max() + 1)  # Normalized
else:  # If no delay columns, set default
    df_fe['Total_delays'] = 0
    df_fe['Total_delays_norm'] = 0

# -----------------------------
# 3️⃣ Credit per product
# -----------------------------
# Debt as a ratio per financial product → risk per product
df_fe['Credit_per_product'] = df_fe['Prct_deuda_vs_ingresos'] / (df_fe['Nro_prod_financieros_deuda'] + 1)
df_fe['Credit_per_product'] = df_fe['Credit_per_product'].replace([np.inf, -np.inf], np.nan).fillna(df_fe['Credit_per_product'].median())

# -----------------------------
# 4️⃣ Has delay flag
# -----------------------------
# Indicates if customer has any late payment
df_fe['Has_delay'] = (df_fe['Total_delays'] > 0).astype(int)

# -----------------------------
# 5️⃣ Debt to Income ratio
# -----------------------------
# Shows proportion of debt relative to income
df_fe['Debt_to_Income'] = df_fe['Prct_deuda_vs_ingresos'] / (df_fe['Mto_ingreso_mensual'] + 1)
df_fe['Debt_to_Income'] = df_fe['Debt_to_Income'].replace([np.inf, -np.inf], np.nan).fillna(df_fe['Debt_to_Income'].median())

# -----------------------------
# 6️⃣ Interaction features
# -----------------------------
# Capture complex relationships that improve model accuracy
df_fe['Income_per_delay'] = df_fe['Income_per_person'] / (df_fe['Total_delays'] + 1)
df_fe['Debt_per_product'] = df_fe['Prct_deuda_vs_ingresos'] / (df_fe['Nro_prod_financieros_deuda'] + 1)
df_fe['Debt_Income_ratio_mult_delay'] = df_fe['Debt_to_Income'] * (df_fe['Total_delays'] + 1)

# -----------------------------
# 7️⃣ Log transformations
# -----------------------------
# Reduce skewness in financial variables
for col in ['Mto_ingreso_mensual','Prct_deuda_vs_ingresos','Income_per_person','Debt_to_Income']:
    df_fe[f'Log_{col}'] = np.log1p(df_fe[col])

```

```python
import pandas as pd
import numpy as np

# Assuming df_fe is your cleaned dataframe
# 5️⃣ Age binning (ensure valid range)
age_min = df_fe['Edad'].min()
age_max = df_fe['Edad'].max()

bins = [age_min - 1, 25, 35, 45, 55, 65, age_max + 1]
labels = ['<=25', '26-35', '36-45', '46-55', '56-65', '65+']
df_fe['Age_bin'] = pd.cut(df_fe['Edad'], bins=bins, labels=labels)


# -----------------------------
# 0️⃣ Fill NaNs without inplace
# -----------------------------
df_fe['Mto_ingreso_mensual'] = df_fe['Mto_ingreso_mensual'].fillna(df_fe['Mto_ingreso_mensual'].median())
df_fe['Prct_deuda_vs_ingresos'] = df_fe['Prct_deuda_vs_ingresos'].fillna(df_fe['Prct_deuda_vs_ingresos'].median())
df_fe['Total_delays'] = df_fe['Total_delays'].fillna(0)

# -----------------------------
# 1️⃣ Income bin (quartiles)
# -----------------------------
df_fe['Income_bin'] = pd.qcut(
    df_fe['Mto_ingreso_mensual'], 
    q=4, 
    labels=['Low','Mid','High','Very High'],
    duplicates='drop'
)

# -----------------------------
# 2️⃣ Debt ratio bin (quartiles)
# -----------------------------
df_fe['Debt_ratio_bin'] = pd.qcut(
    df_fe['Prct_deuda_vs_ingresos'], 
    q=4, 
    labels=['Low','Mid','High','Very High'],
    duplicates='drop'
)

# -----------------------------
# 3️⃣ Delay bin (custom cut)
# -----------------------------
max_delay = df_fe['Total_delays'].max()

# Ensure last bin is strictly larger than previous one
last_bin = max(max_delay + 1, 11)  # at least 11 if previous bin was 10
bins = [-1, 0, 1, 3, 10, last_bin]

df_fe['Delay_bin'] = pd.cut(
    df_fe['Total_delays'],
    bins=bins,
    labels=['None','Low','Medium','High','Very High'],
    include_lowest=True
)
# -----------------------------
# 4️⃣ One-hot encoding
# -----------------------------
df_fe = pd.get_dummies(
    df_fe,
    columns=['Income_bin','Debt_ratio_bin','Delay_bin'],
    drop_first=True
)

# -----------------------------
# 5️⃣ Verify
# -----------------------------
print(df_fe.head())
print("Shape after one-hot encoding:", df_fe.shape)
```
